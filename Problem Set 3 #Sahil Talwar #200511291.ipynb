```python
#step 1
import pandas as pd
import datetime

#step 2 & 3
data = pd.read_csv("wind.txt", sep = "\s+", parse_dates = [[0,1,2]])

data
```

```python
#step 4


# function that uses datetime
def fix_century(x):
  year = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(year, x.month, x.day)

# apply the function fix_century on the column and replace the values to the right ones
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)

# data.info()
data.head()
```

```python
def date_time(d):
    year = d.year - 100 if d.year > 1989 else d.year
    return datetime.date(year, d.month, d.day)
    
data['Yr_Mo_Dy']  = data['Yr_Mo_Dy'].apply(date_time)

data
```

```python
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy')
data
```

# Question 6
Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
data.head()
```

```python
marriages_capita = data['Marriages_per_1000']
divorces_capita = data['Divorces_per_1000']

columns = marriages_capita, divorces_capita
years = data['Year']
ind = ["marriages per capita", "divorces per capita"]

x_data = range(1867, 1867 + data.shape[0])

fig, axis = plt.subplots()


# plot each column
for i in range(len(columns)):
    axis.plot(x_data, columns[i], label=ind[i])

    
axis.set_xlabel('Year')
axis.set_ylabel('Marriage/Divorces per 1000 people')
axis.legend()
```

# Question 7 
Create a vertical bar chart comparing the number of marriages and divorces per-capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("us-marriages-divorces-1867-2014.csv")


marriages = data['Marriages_per_1000']
divorces = data['Divorces_per_1000']

year = [1900, 1950, 2000]
ind = ["marriages per capita", "divorces per capita"]

da = data.loc[data['Year'] == 1900]
da = da.append(data.loc[data['Year'] == 1950])
da = da.append(data.loc[data['Year'] == 2000])
columns = da['Marriages_per_1000'], da['Divorces_per_1000']

fig, p = plt.subplots()
bar_width = 0.40
X = np.arange(3)


p = plt.bar(X, columns[0], bar_width, label="Marriages count", color='r')
p = plt.bar(X + bar_width, columns[1], bar_width, label="Divorces count")

plt.xlabel('Year')
plt.ylabel('Marriage/Divorces per capita')
plt.xticks(X + (bar_width/2) , ("1900", "1950", "2000"))
plt.legend()
plt.tight_layout()
```

# Question 10
Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year.

https://jakevdp.github.io/PythonDataScienceHandbook/04.02-simple-scatter-plots.html

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.pyplot as plt
%matplotlib inline


arcade_revenue_data = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
arcade_revenue = arcade_revenue_data['Total Arcade Revenue (billions)'].values
cs_doctorates = arcade_revenue_data['Computer Science Doctorates Awarded (US)'].values


fig, p = plt.subplots()

colors = cm.rainbow(np.linspace(0, 1, len(arcade_revenue_data['Year'])))

for i in range(len(arcade_revenue_data['Year'])):
    p.scatter(arcade_revenue[i], cs_doctorates[i],color=colors[i])

p.set_xlabel('Total Revenues Earned by Arcades')
p.set_ylabel('Number of Computer Science PhDs Awarded (US)')
plt.show()
```
